```python
import pywt
#baca data sinyal eeg
filename = "Bader_Autism_24_11_2011S001R01.dat"
with b2k.BCI2kReader(filename) as test:
        my_signals, stateslice = test[0:200]

my_data = np.fft.fft2(my_signals)
coeffs = pywt.dwt2(my_data, 'haar')
cA, (cH, cV, cD) = coeffs

my_data2 = cA.astype(np.uint8)
plt.imshow(my_data2)
```

# With FFT & DWT Autism

```python
import pywt
import scipy as sp
from scipy import signal

#baca data sinyal eeg
filename = "Mada_Autism_26_5_2011S001R01.dat"
with b2k.BCI2kReader(filename) as test:
        my_signals, stateslice = test[0:1000]
        
coeffs = pywt.dwt2(my_signals, 'haar')
cA, (cH, cV, cD) = coeffs

my_data2 = cA.astype(np.uint8)

plt.figure(figsize=(100,10))    
plt.pcolormesh(my_data2, cmap='jet')
plt.yscale('linear')
plt.axis('off')
plt.savefig('normal_dwt2.png', transparent = True, bbox_inches='tight')
```

```python
import pywt
import scipy as sp
from scipy import signal
from BCI2kReader import BCI2kReader as b2k
import numpy as np
import matplotlib.pyplot as plt

#baca data sinyal eeg
filename = "Dhelal_Normal_15_6_2011S001R02.dat"
with b2k.BCI2kReader(filename) as test:
        my_signals, stateslice = test[0:1000]
        
coeffs = pywt.wavedec(my_signals, 'dmey', level=8) #DWT
cA8, cD8, cD7, cD7, cD5, cD4, cD3, cD2, cD1 = coeffs

my_data2 = cA8.astype(np.uint8)

plt.pcolormesh(my_data2, cmap='jet')
plt.yscale('linear')

plt.xlabel('Time (s)')
plt.ylabel('Freq (Hz)')
```
