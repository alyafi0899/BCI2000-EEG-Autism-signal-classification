```python
import pywt
from BCI2kReader import BCI2kReader as b2k
import scipy as sp
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

#baca data sinyal eeg
filename = "./dataset - Copy/normal/Amer_Normal_5_5_2011S001R01.dat"
with b2k.BCI2kReader(filename) as test:
        my_signals, stateslice = test[0:1000]
        
coeffs = pywt.dwt2(my_signals, 'haar')
cA, (cH, cV, cD) = coeffs

my_data2 = cA.astype(np.uint8)

plt.figure(figsize=(100,10))    
plt.pcolormesh(my_data2, cmap='jet')
plt.yscale('linear')
plt.axis('off')
plt.savefig('normal_dwt2.png', transparent = True, bbox_inches='tight')
```

```python
import pywt
from BCI2kReader import BCI2kReader as b2k
import scipy as sp
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

#baca data sinyal eeg
filename = "./dataset - Copy/normal/Omran_Normal_5_5_2011S001R01.dat"
with b2k.BCI2kReader(filename) as test:
        my_signals, stateslice = test[0:1000]

scales=np.arange(1, 1001)
cwtmatr, freqs = pywt.cwt(my_signals, scales, 'mexh')

plt.imshow(cwtmatr, cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())  
plt.show() 
```
